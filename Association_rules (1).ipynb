```python
# Data Preprocessing
# Drop rows with missing values
# Check the actual column names in your DataFrame using df.columns
print(df.columns)  # Print the column names to verify
# Assuming the column name is "Product" and you want to drop rows where it is NaN
df.dropna(subset=['Product'], inplace=True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Filter necessary columns (Product)
df = df[['Product']]
```

```python
# Create a basket where products are organized by transaction (one-hot encoded)
# Create a dummy transaction identifier:
df['TransactionID'] = df.groupby('Product').cumcount()+1
# Create the basket using crosstab:
basket = pd.crosstab(df['Product'], df['TransactionID'])

# basket now contains products as index and transaction IDs as columns.
# Each cell value is 1 if the product was in the transaction, 0 otherwise.
```

```python
# Convert quantity to a binary indicator (1 if purchased, 0 otherwise)
basket = basket.map(lambda x: 1 if x > 0 else 0)
```

```python
# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
```

```python
# Generate the association rules with confidence and lift
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
```

```python
# Filter for rules with high confidence and lift
rules = rules[(rules['confidence'] >= 0.5) & (rules['lift'] >= 1.2)]
```

```python
# Filter rules based on lift
rules = rules[rules['lift'] > 1]
```

```python
# Display the rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
# Save rules to a CSV for further analysis
rules.to_csv('association_rules.csv', index=False)
```
